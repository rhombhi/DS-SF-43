```python

path = "../../data/loansData.csv"

loans = pd.read_csv(path)

loans.head()
```

###
1. make lowercase, convert dots to _
2. remove % from int rate
3. rmove months from loan length
0. replace nulls by 'unknown'
4. remove teh % for Debt.To.Income.Ratio
5. split FICO range into fico min and fico max
6. remove years from the emp len

```python
#1. make lowercase, convert dots to _
loans.columns = loans.columns.str.lower()
```

```python
loans.columns
```

```python
#1. make lowercase, convert dots to _
loans.columns = loans.columns.str.replace('.','_')
loans.columns
```

```python
#2. remove % from int rate

loans.interest_rate = loans.interest_rate.str.replace('%','')
loans.interest_rate.head()
```

```python
#4. remove teh % for Debt.To.Income.Ratio
loans.debt_to_income_ratio = loans.debt_to_income_ratio.str.replace('%','')
loans.debt_to_income_ratio.head()
```

```python
loans.dtypes
```

```python
loans.debt_to_income_ratio = loans.debt_to_income_ratio.astype('float64')
loans.interest_rate = loans.interest_rate.astype('float64')
```

```python
loans.dtypes
```

```python
#3. rmove months from loan length

# loans.loan_length.head()

loans.loan_length = loans.loan_length.str.replace('months','')
loans.loan_length = loans.loan_length.astype('int64')

loans.loan_length.head()
```

```python
# 5. split FICO range into fico min and fico max

loans.fico_range.head()
```

```python

# loans.fico_range.str.slice(0,3)
# loans.fico_range.str.slice(4,7)
loans['fico_min'] = loans.fico_range.str.slice(0,3)
loans['fico_max'] = loans.fico_range.str.slice(4,7)
```

```python
loans.fico_min.head()
```

```python
loans.dtypes
```

```python
loans.fico_min = loans.fico_min.astype('int64')
loans.fico_max = loans.fico_max.astype('int64')
```
